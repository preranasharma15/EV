PERFORMING SENTIMENT ANALYSIS OF THE REVIEWS

```python
nltk.download('vader_lexicon')
```

```python
sentiments = SentimentIntensityAnalyzer() 
sentiments_list = []

for review_text in h['review']:
    if pd.isna(review_text):
        sentiments_list.append('neutral')
        continue
    sentiment_scores = sentiments.polarity_scores(review_text)

    positive_score = sentiment_scores['pos']
    negative_score = sentiment_scores['neg']

    if positive_score > negative_score:
        sentiment_label = 'positive'
    elif positive_score < negative_score:
        sentiment_label = 'negative'
    else:
        sentiment_label = 'neutral'

    sentiments_list.append(sentiment_label)

h['sentiment'] = sentiments_list
```

```python
h['sentiment'].value_counts()
```

```python
data = h['sentiment'].value_counts()
labels = ['Positive', 'Negative', 'Neutral']
explode = [0.05, 0.05, 0.1]
plt.figure(figsize=(6,6))
plt.pie(data, explode=explode, labels=labels, shadow=True, autopct='%.2f%%')
plt.title('Sentiment Analysis of Reviews', fontsize=20)
plt.legend(['positive','negative', 'neutral'])
plt.show()
```

```python
data_segment = h[['Used it for', 'Owned for', 'Ridden for', 'rating',
       'Visual Appeal', 'Reliability', 'Performance', 'Service Experience',
       'Extra Features', 'Comfort', 'Maintenance cost', 'Value for Money',
       'sentiment']].copy()
```

```python
data_segment = h.iloc[:, 5:13].copy()
```

```python
data_segment = data_segment.fillna(0.)
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_segment)
```

```python
pca = PCA(random_state = 42)
pca.fit(data_scaled)
```

```python
data_pca = pca.transform(data_scaled)
h_pca = pd.DataFrame(data_pca, columns = [f'PC{x +1}' for x in range(len(data_segment.columns))])
pca_summary = pd.DataFrame({'Standard Deviation':h_pca.std(),
                           'Proportion of Variance': pca.explained_variance_ratio_,
                           'Cumultive Proportion': pca.explained_variance_ratio_.cumsum()})
pca_summary
```

```python
factor_loadings = pd.DataFrame(pca.components_, columns = data_segment.columns, index = h_pca.columns).T
```

```python
factor_loadings.style.background_gradient(cmap = 'Blues')
```

```python
index_names = factor_loadings.index

sns.scatterplot(data=h_pca, x = 'PC1', y = 'PC2', color = ['orange'])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA of the EV 2-Wheeler dataset')
plt.grid(False)
plt.show()
```

```python
data_km28 = [KMeans(n_clusters = k, random_state = 42).fit(data_scaled) for k in range(2, 9)]
wcss = [model.inertia_ for model in data_km28]
fig = plt.figure(figsize = (6,4))
sns.barplot(x = list(range(2, 9)), y = wcss, color = 'pink')
plt.xlabel("Segments")
plt.ylabel('Sum of Within Cluster Distances')
plt.title("Graph for the EV 2-Wheeler data set")
plt.show()
```

```python
kmeans = data_km28[2]
data_clust = pd.DataFrame(data_scaled, columns = data_segment.columns)
data_clust['cluster'] = kmeans.labels_
data_pca = h_pca.copy()
data_pca['cluster'] = kmeans.labels_
data_profile = data_segment.copy()
data_profile['cluster'] = kmeans.labels_
columns_segment = data_clust.columns
data_pivot = data_profile[columns_segment].groupby('cluster').mean().T
data_pivot_mean = data_pivot.mean(axis = 1).reset_index()
data_pivot_mean.columns = ['Variable', 'Value']
```

```python
plt.figure(figsize = (12, 9))
for i in range(4):
    plt.subplot(2, 2, i+1)
    sns.barplot(data=data_pivot, x = i, y = data_pivot.index)
    sns.scatterplot(data=data_pivot_mean, x = 'Value', y = 'Variable', color = 'black')
    for index, row in data_pivot_mean.iterrows():
        plt.hlines(y=row['Variable'], xmin=0, xmax=row['Value'], colors='black', alpha = 0.2)
    plt.ylabel("")
    plt.xlabel("")
    plt.title(f"Cluster {i}: {data_profile['cluster'].value_counts()[i]} ({data_profile['cluster'].value_counts()[i]*100/len(data_profile):.0f}%)")
plt.suptitle("Segment profile plot for the four-segment solution for the EV 2-Wheeler dataset", fontsize = 15)
plt.tight_layout()
plt.savefig("segment_profile.png")
plt.show()
```

The above graph plot of different customers requirements that Segmnet 0 values appearence, segment 1 doesn't care aout anything and seems displeased, segment 2 values features of the vehicle and segment 3 has main focus on comfort.

```python
data_desc = h.copy()
data_desc['cluster'] = kmeans.labels_
def mosaicplot(df, color_list, figsize=(5,5), barwidth = 1, title = "Simple Mosaic Plot", xlabel = 'cat'):

    fig, ax = plt.subplots(figsize =figsize)

    # Convert to ratio of each row sum and perform a cumsum along row
    p_cnt_ratio_df = df.div(df.sum(1), axis =0)
    ratio_cum_df = p_cnt_ratio_df.cumsum(axis=1)

    # column order.
    col_order = ratio_cum_df.columns.tolist()

    x = [str(n) for n in ratio_cum_df.index.tolist()]

    
    for col, colour in zip(col_order[::-1], color_list):
        plt.bar(x, ratio_cum_df[col], color=colour, edgecolor='white', width=barwidth, label=col)

    # Add in legend
    plt.legend(loc='upper left', bbox_to_anchor=(1,1), ncol=1)


    # Configure title and axis
    plt.title(title)
    plt.xlabel("Segments")
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)
    plt.yticks([])
    plt.show()
color_list = ['#FF90BC','#EEC759', '#59B4C3', '#FA7070', '#ADD899', '#BC7AF9', '#B5CB99', '#FF8080']
sentiment_cluster = pd.crosstab(index = data_desc['cluster'], columns = data_desc['sentiment'])
sentiment_cluster
```

```python
mosaicplot(sentiment_cluster, color_list[:3], figsize=(6,5), barwidth = 1)
```

It is clearly evident that Segment 0, Segment 1 and  Segment 3 shows more positive comments whereas Segment 2 shows more negative than positive.
